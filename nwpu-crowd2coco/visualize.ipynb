```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import json

img = cv2.imread('./images/0033.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
info = json.load(open('./jsons/0033.json', 'r'))
boxes = info['boxes']

for box in boxes:
    x1, y1, x2, y2 = int(box[0]), int(box[1]), int(box[2]), int(box[3])
    img = cv2.rectangle(img, (x1,y1), (x2,y2), (0,255,0), 2)
plt.imshow(img)
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import json
import os

# img = cv2.imread('../coco/val2017/000000289343.jpg')
# img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
info = json.load(open('./annotations/nwpu-crowd_train.json', 'r'))
annotations = info['annotations']
images = info['images']
```

```python
print(images[0])
print(annotations[0])
file_name = images[0]['file_name']
img = cv2.imread(os.path.join('./cat', file_name))
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
for annotation in annotations:
    bbox = annotation['bbox']
    x1, y1, x2, y2 = int(bbox[0]), int(bbox[1]), int(bbox[0]+bbox[2]), int(bbox[1]+bbox[3])
    img = cv2.rectangle(img, (x1,y1), (x2,y2), (0,255,0), 2)
plt.imshow(img)
plt.show()
```
